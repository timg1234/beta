# GET Forms

```python
import requests
from bs4 import BeautifulSoup
```

```python
content = requests.get("https://www.imdb.com/find?q=Mark&s=nm&exact=true&ref_=fn_nm_ex")
```

```python
doc = BeautifulSoup(content.text, 'html.parser')
```

```python
# Grad all of the titles
name_tags = doc.find_all(class_="result_text")

# Let's print the first 5
for name in name_tags:
    print(name.a.string)
```

```python
data = {
    'lastName': 'Kelly',
    'firstName': 'John',
    'boardCode': '9',
    'licenseNumber': '',
    'licenseType': '250',
    'registryNumber': ''
}

url = "https://search.dca.ca.gov/results"
```

```python
response = requests.post(url, data=data)
doc = BeautifulSoup(response.text, 'html.parser')

# Grabbing all the name tags
name_tags = doc.find_all("h3")

# Let's print the first 5
for name in name_tags:
    print(name.text)
```

# Cookies & Button Input Values to an URL

```python
# Scraping Sites with Logins
import requests
from lxml import html
from bs4 import BeautifulSoup
```

```python
# login and to-scrape URL
login_url = "http://quotes.toscrape.com/login"
get_url = "http://quotes.toscrape.com/"
```

```python
# session object
session_req = requests.session()
```

```python
# getting the CSRF token
login = session_req.get(login_url)
login_html = html.fromstring(login.content)
authenticity_token = list(set(login_html.xpath("//input[@type='hidden']/@value")))[0]
print(authenticity_token)
```

```python
payload = {
    "username": "yo",
    "password": "password",
    "csrf_token": authenticity_token
}
```

```python
# POST request
post = session_req.post(
    login_url,
    data = payload
)
```

```python
# Getting the required page
result = session_req.get(
    get_url
)
```

```python
soup = BeautifulSoup(result.text, "html.parser")
soup
```
